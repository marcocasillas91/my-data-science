```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup
```

```python
standings_url= 'https://fbref.com/en/comps/31/Liga-MX-Stats'
data = req.get(standings_url)
```

```python
soup = BeautifulSoup(data.text, features = "html.parser")
```

```python
#Select table in soup
standings_table=soup.select('table.stats_table')[0]
```

```python
# Use command find_all to retrieve team links
links = standings_table.find_all('a')
# type(links)  
#links is a beautifulsoup result set
```

```python
links =  [l.get("href") for l in links]
```

```python
links = [l for l in links if '/squads/' in l]
```

```python
links
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```

```python
chivas_url = team_urls[6]
```

```python
chivas_data = req.get(chivas_url)
```

```python
type(team_urls)
```

```python
# In order to use read_html, lxml should have been installed. 
chivasMatches= pd.read_html(chivas_data.text, match = "Scores & Fixtures")[0]
```

```python
chivasMatches.head()
```

```python
# Soup to retrieve team's shotting imformation
soup_1 = BeautifulSoup(chivas_data.text, features = "html.parser")
shooting_link = soup_1.find_all("a")
shooting_link = [l.get("href") for l in shooting_link]
shooting_link = [l for l in shooting_link if l and 'all_comps/shooting/' in l]
shooting_link
```

```python
shooting_data = req.get(f"https://fbref.com{shooting_link[0]}")
shooting_data
```

```python
shootingInfo = pd.read_html(shooting_data.text, match='Shooting')[0]
shootingInfo.head()
```

```python
shootingInfo.columns = shootingInfo.columns.droplevel()
```

```python
chivasMatches.head()
```

```python
shootingInfo.head()
```

```python
team_data = chivasMatches.merge(shootingInfo[["Date","Sh","SoT", "Dist","FK","PK","PKatt"]], on = "Date")
```
